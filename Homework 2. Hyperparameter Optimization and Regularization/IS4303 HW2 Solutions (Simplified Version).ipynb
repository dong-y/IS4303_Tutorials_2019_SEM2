```python
lasso = LogisticRegression(fit_intercept=True, penalty='l1', solver='liblinear', max_iter=1000, tol=0.01, random_state=12345)

# Create N regularization strength paramaters: lambdas
N = 500
lbd = np.logspace(-6, 6, N)
lbd
```

```python
coefs = [] # Dimension: N*1*70

# Train the model with different regularisation strengths
for a in lbd:
    lasso.set_params(C=1/a) # Note that C is inverse value of regularization strength parameter "lambda"
    lasso_model = lasso.fit(X_train, y_train)
    coefs.append(lasso_model.coef_)
    
coefs_array = np.array(coefs)
coefs_array = coefs_array.reshape((N, 70))
print(coefs_array)
```

```python
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(np.log10(lbd), coefs_array)
plt.xlabel('log10(lambda)')
plt.ylabel('Coefficients')
plt.title('Lasso coefficients v.s. Regularization Strength Parameters (lambda)')
plt.axis('tight')
plt.show()
```

<div class="alert alert-danger">
<b>Findings:</b><p>As regularization strength parameter increases, coefficient size will quickly/suddenly drop to 0. We penalize more on large-size coefficients. That is how Lasso regression does feature selection (i.e., Embedded method).</p>
</div>

```python
# How does accuracy and auc change with lambda value?
import plotly as py
import plotly.graph_objs as go
import ipywidgets as widgets
from sklearn.metrics import roc_curve, auc
py.offline.init_notebook_mode(connected=True)
```

```python
# How does accuracy and auc change with lambda value?
def update_plot(a):
    
    lasso = LogisticRegression(fit_intercept=True, penalty='l1', solver='liblinear', C=1/a, random_state=12345)
    lasso_model = lasso.fit(X=X_train, y=y_train)
    y_pred = lasso_model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    scores = lasso_model.predict_proba(X_test)[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test, scores) # fpr: FP Rate, tpr: TP Rate, thresholds: Pr(y=1)
    roc_auc = auc(fpr, tpr)
    
    layout = go.Layout(
        title = 'Receiver operating characteristic of Lasso model',
        xaxis = {
            'title': 'False Positive Rate',
            'range': [0, 1.1]
        },
        yaxis = {
            'title': 'True Positive Rate',
            'range': [0, 1.1]
        }
    )

    trace_reference = go.Scatter(
        x = [0, 1],
        y = [0, 1],
        mode = 'lines',
        name = 'Reference Line',
        line = {
            'shape': 'spline',
            'dash': 'dashdot',
            'color': 'red',
            'width': 2
        }    
    )
    
    trace_roc = go.Scatter(
        x = fpr,
        y = tpr,
        mode = 'lines',
        name = 'lambda=%0.2f, \n Accuracy=%f,\n AUC=%f' %(a, accuracy, roc_auc),
        line = {
            'shape': 'spline'
        }
    )  

    fig = go.Figure(data=[trace_roc, trace_reference], layout=layout)
    py.offline.iplot(fig)    
```

```python
lambdas = widgets.FloatLogSlider(
    value = 1,
    base = 10,
    min = -6, # max exponent of base
    max = 6, # min exponent of base
    step = 0.1, # exponent step
    description = 'lambdas: '
)

print("Please drag and slide: \n")
widgets.interactive(update_plot, a=lambdas)
```

<a id="Q5"></a>

<div class="alert alert-info">
<b>Question 5 (1 point): From Question 4, Find and Select The Best Regularization Strength $\lambda$ That Provides Highest Test Accuracy For Lasso Model.</b> 
<div>Please report/print:
   <ol>
    <li>This best $\lambda$;</li>
    <li>Its corresponding test accuracy.</li>
   </ol>
</div>
</div>

```python
# Create N regularization strength paramaters: lambdas
N = 500
lbd = np.logspace(-6, 6, N)

# Train the model with different lambdas, and find the best lambda
lasso_max_accuracy = 0
lasso_best_lambda = 0
lasso_best_coefficient = np.array([])

for a in lbd:
    lasso.set_params(C=1/a) # Note that C is inverse value of regularization strength parameter "lambda"
    lasso_model = lasso.fit(X_train, y_train) # Fit Lasso model   
    y_pred = lasso_model.predict(X_test) # Prediction
    accuracy = accuracy_score(y_test, y_pred) # Test accuracy
    
    if accuracy >= lasso_max_accuracy:
        lasso_max_accuracy = accuracy
        lasso_best_lambda = a
        lasso_best_coefficient = lasso_model.coef_[0]
```

```python
print("Best lambda is {} and highest test accuracy is {}.".format(lasso_best_lambda, lasso_max_accuracy))
print("Lasso coefficients of the best lambda are: \n {}".format(lasso_best_coefficient))
```

```python
ridge = LogisticRegression(fit_intercept=True, penalty='l2', solver='liblinear', max_iter=1000, tol=0.01, random_state=12345)

# Create N regularization strength paramaters: lambdas
N = 500
lbd = np.logspace(-1, 11, N)
lbd
```

```python
coefs = [] # Dimension: N*1*70

# Train the model with different regularisation strengths
for a in lbd:
    ridge.set_params(C=1/a) # Note that C is inverse value of regularization strength parameter "lambda"
    ridge_model = ridge.fit(X_train, y_train)
    coefs.append(ridge_model.coef_)
    
coefs_array = np.array(coefs)
coefs_array = coefs_array.reshape((N, 70))
print(coefs_array)
```

```python
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(np.log10(lbd), coefs_array)
plt.xlabel('log10(lambda)')
plt.ylabel('Coefficients')
plt.title('Ridge coefficients v.s. Regularization Strength Parameters (lambda)')
plt.axis('tight')
plt.show()
```

<div class="alert alert-danger">
<b>Findings:</b><p>As regularization strength parameter increases, coefficient size will slowly decrease, get closer and closer to 0 (but may not be exactly 0). Again, we penalize more on large-size coefficients, but Ridge regression does not really do feature selection.</p>

```python
# Create N regularization strength paramaters: lambdas
N = 500
lbd = np.logspace(-1, 11, N)

# Train the model with different lambdas, and find the best lambda
ridge_max_accuracy = 0
ridge_best_lambda = 0
ridge_best_coefficient = np.array([])

for a in lbd:
    ridge.set_params(C=1/a) # Note that C is inverse value of regularization strength parameter "lambda"
    ridge_model = ridge.fit(X_train, y_train) # Fit Ridge model   
    y_pred = ridge_model.predict(X_test) # Prediction
    accuracy = accuracy_score(y_test, y_pred) # Test accuracy
    
    if accuracy >= ridge_max_accuracy:
        ridge_max_accuracy = accuracy
        ridge_best_lambda = a
        ridge_best_coefficient = ridge_model.coef_[0]
```

```python
print("Best lambda is {} and highest test accuracy is {}.".format(ridge_best_lambda, ridge_max_accuracy))
print("Ridge coefficients of the best lambda are: \n {}".format(ridge_best_coefficient))
```
